### Analyse d'un Magasin de Téléphones

Un gérant d'un grand-Magasin de téléphone, fait appel à votre expertise en tant que **Data Analyst**, pour connaître __quel modèle de téléphone, peut-il facilement écouler de son stocks.__ Alors quel est selon vous le modèle qui pourra être facilement écouler sur le marché ?

Vous allez trouver la base de données sur ce site : https://www.kaggle.com/datasets/juanmerinobermejo/smartphones-price-dataset

La base de données contient les informations :

- `Smartphone Name :` The unique identifier or model name of the smartphone/_L'identifiant unique ou le nom du modèle du smartphone._
- `Brand :` Smartphone brand/_Marque de téléphone._
- `Model :` Smartphone brand model/_Modèle de marque de téléphone._
- `RAM (Random Access Memory) :` The amount of memory available for multitasking/_La quantité de mémoire disponible pour les multitâches._
- `Storage :` capacity of the smartphone/_Capacité du téléphone._
- `Color :` color of the smartphone/_couleur du téléphone._
- `Free :` Yes/No if the smartphone is attached to a cell company contract/_Oui/Non si le smartphone est rattaché à un contrat d'entreprise de téléphonie mobile._
- `Price :` The cost of the smartphone in the respective currency/_Le coût du smartphone dans la devise respective._

_En utilisant cet ensemble de données, les chercheurs et les analystes peuvent explorer les modèles, les tendances et les relations entre les spécifications des smartphones et leur prix._ _Il constitue une excellente ressource pour des tâches telles que la prévision des prix, l'analyse du marché et la comparaison de différentes configurations de smartphones._ _Que vous souhaitiez identifier les options les plus rentables ou comprendre l’impact de composants matériels spécifiques sur les prix des smartphones, cet ensemble de données offre de nombreuses possibilités d’exploration approfondie._

## Compter le nombre de modèles et de téléphones par Marque de fabricants

### Nombres de modèle par Fabricant

```python
# Nombre de modèles 
models = telephone[['Brand', 'Model']].groupby('Brand').nunique().reset_index() # Compter le Total des valeurs uniques 
models.rename(columns= {'Brand': 'Fabricant', 'Model': 'Nombre_Model'}, inplace= True) # Remplacer le nom d'une colonne
model_par_fabricant= models.sort_values('Nombre_Model',ascending= False) # Classer par ordre numérique où Alphabétique (Ici, nous ferons un classement par ordre décroissant)
model_par_fabricant
```

```python
# Représentation grapgique des leaders (10 premiers)
top_modeles = model_par_fabricant[0:10]
plt.bar(top_modeles['Fabricant'],top_modeles['Nombre_Model'], color= 'skyblue')
plt.title("Nombres de Modèles par Fabricant") # Titre du graphique
plt.xlabel("Fabricant") # Titre de l'axe des x
plt.ylabel("Nombres de Marques") # Titrede l'axe des y
for i, freq in enumerate(top_modeles['Nombre_Model']):
    plt.text(i,freq,str(freq),ha= 'center',va= 'bottom')
plt.xticks(rotation= 45) # orientation des valeurs de x
plt.figure(figsize=(20,30)) # Ajuster la Largeur, et Hauteur du graphique.
plt.show()
```

```python
nombre_tel = telephone['Brand'].value_counts().reset_index()# Compter le Total des valeurs uniques 
nombre_tel.rename(columns= {'Brand': 'Fabricant', 'count': 'Total_Telephone'}, inplace= True) # Remplacer le nom d'une colonne
telephone_par_fabricant= nombre_tel.sort_values('Total_Telephone',ascending= False) # Classer par ordre numérique où Alphabétique (Ici, nous ferons un classement par ordre décroissant)
telephone_par_fabricant
```

```python
# Représentation grapgique (bar_Horizontal)

learder_fabricant = telephone_par_fabricant[0:10]

plt.bar(learder_fabricant['Fabricant'],learder_fabricant['Total_Telephone'], color = 'blue')
plt.title("Nombre de Telephone par Fabricant") # Titre du graphique
plt.xlabel("Fabricant") # Titre de l'axe des x
plt.ylabel("Nombres de Telephone") # Titre de l'axe des y
for i,freq in enumerate(learder_fabricant['Total_Telephone']):
    plt.text(i,freq, str(freq), ha= 'center', va='bottom')
plt.xticks(rotation= 45) # orientation des valeurs de X
plt.figure(figsize=(12,25)) # Ajuster la Largeur, et Hauteur du graphique.
plt.show()
```

### Quelle est la corrélation entre le Prix d'un telephone et sa capacité interne (Storage), et sa vitesse d'exécution (RAM) ?

### Corrélation entre *le Stockage Interne* et le *Prix du téléphone*

```python
# Représentation graphique .

prix = [telephone['Final Price']]
stockage = [telephone['Storage']]

plt.scatter(stockage,prix, color= 'green')
plt.title("Stockage_Interne et Prix des Telephones",fontdict={'weight': 'bold', 'style': 'italic', 'size': 16, 'color': 'black'})
plt.xlabel("Stockage Interne")
plt.ylabel(" Prix")
```

### Corrélation entre le prix du telephone et sa vitesse d'execution (RAM)

```python
prix = telephone['Final Price']
ram = telephone['RAM']

plt.scatter(ram,prix, color= 'blue')
plt.title("RAM et Prix des Telephones", fontdict={'weight': 'bold', 'style': 'italic', 'size': 16, 'color': 'black'})
plt.xlabel("Vitesse d'execution (RAM)")
plt.ylabel("Prix")
```

### Commentaire sur les différentes corrélations.

##### Nous pouvons consater, qu'il n'y a aucune corrélation sur les deux graphiques, plutôt des dispersions.Car comme nous pouvons le constater sur le:
- Premier _1er_ grapghique (_Corrélation Prix_Stockage_), qu'un Telephone qui une capacité interne de *256 Go coûte plus de 2000 dollars, qu'un téléphone de 512  Go qui coûte moins de  2000 dollars.* 
- Deuxième _2ème_ graphique (_Corrélation Prix_RAM_), un téléphone qui à une RAM de *6 Go coûte plus de 2000 dollars, qu'un telephone ayant une RAM de 12 Go, coûte moins de 2000 dollars*.
